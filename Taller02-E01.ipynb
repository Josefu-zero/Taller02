# **EJERCICIO 1** 
Encuentra todas las ra√≠ces del polinomio:

$$
x^4 + 540x^3 + 109124x^2 + 9781632x + 328188672 = 0
$$



```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Define the parabola equation: y = ax^2 + bx + c
a, b, c ,d, e = 1, 540, 109124, 9781632, 328188672  # Example coefficients
x = np.linspace(-154, -123, 1000)  # Generate x values 
y = (a * x**4 + b * x**3 + c * x**2 + d * x + e)  # Compute y values
coefficients = [a, b, c, d,e]  # Coefficients of the polynomial
roots = np.roots(coefficients)

from matplotlib.animation import FuncAnimation

# Configurar la figura y los ejes
fig, ax = plt.subplots()
ax.set_xlim(x.min(), x.max())
ax.set_ylim(y.min(), y.max())
ax.axhline(0, color='black', linewidth=0.5, linestyle='--')
ax.axvline(0, color='black', linewidth=0.5, linestyle='--')
ax.set_title('Iteraciones para encontrar ra√≠ces')
ax.set_xlabel('x')
ax.set_ylabel('y')

# L√≠nea de la funci√≥n y puntos de las ra√≠ces
line, = ax.plot(x, y, label='Polinomio', color='blue')
points, = ax.plot([], [], 'ro', label='Ra√≠ces aproximadas')

# Inicializar la animaci√≥n
def init():
    points.set_data([], [])
    return points,

# Actualizar la animaci√≥n en cada iteraci√≥n
def update(frame):
    current_roots = np.roots(coefficients[:frame + 1])  # Calcular ra√≠ces aproximadas
    points.set_data(current_roots.real, current_roots.imag)
    return points,

# Crear la animaci√≥n
ani = FuncAnimation(fig, update, frames=len(coefficients), init_func=init, blit=True, interval=500)

# Mostrar la animaci√≥n
plt.legend()
plt.show()
# Guardar la animaci√≥n como un archivo GIF
ani.save('animationBiseccionE1.gif', writer='pillow', fps=2)
```
